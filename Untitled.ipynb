```python
import Loader
from PreProcessor import Preprocessor
from LdaMalletHandler import LdaMalletHandler
from DocumentRetriever import DocumentRetriever
from gensim.corpora.dictionary import Dictionary
from gensim.models.wrappers import LdaMallet
import numpy as np
```

```python
multi_model_name = "demo"
from_n = 50
to_n = 200
interval = 10
```

```python
corpus = Loader.load_from_folder("data")
p = Preprocessor(lang='portuguese')  
pre_processed = p.transform(corpus['articles'])
```

```python
tp_range = np.arange(from_n,to_n+1,interval,dtype=np.uint32)
```

```python
for n_topics in tp_range:
    name = multi_model_name+"_"+str(n_topics)+"_topics"
    lda = LdaMalletHandler(mallet_path='/Users/alex/mallet-2.0.8/bin/mallet')
    lda.run_model(model_name=name, corpus=pre_processed, num_topics=n_topics, iterations=100)
    file = open("results/"+name+".txt", "w")
    for topic in lda.get_string_topics():
        file.write(topic+"\n")
    file.close()    
    lda.save_model()
```
