```python
import sys
from scipy.stats import spearmanr
import requests
import numpy as np
from os import listdir
from datetime import datetime
from os.path import isdir, isfile, join
```

```python
host = "http://127.0.0.1:5000"

def rank_topics(path):
    file = open(path).read();
    lines = file.splitlines()
    topics = []
    i=0
    for topic in lines:
        r = requests.post(host+"/checktopic/", data={'topic':" ".join(topic.split()), 'window_size':10}) 
        if (r.status_code == 200 and eval(r.text)["code"] == 200 ):
            topics.append([topic, eval(r.text)["pmi"], i])
        
        topics.sort(key=lambda x: x[1], reverse=True)
        i+=1
    return topics
```

```python
ranked = rank_topics("results/FolhaCompleto100t_1000i.txt")
```
